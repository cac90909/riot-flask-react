> Note, in order to actually retrieve the champion name, instead of just their ID, you have to use this JSON page. This is because the 'match detail' objects contain a key called 'participants'. Each participant has a unique 'championId' which is identical to the 'key' parameter from the JSON given on: http://ddragon.leagueoflegends.com/cdn/11.11.1/data/en_US/champion.jsonhttp://ddragon.leagueoflegends.com/cdn/11.11.1/data/en_US/champion.json. Simply looking up the key from this JSON link and connecting it to the championId allows us to display actual champion names instead of just ids. Additionally, we can use the name to connect to: http://ddragon.leagueoflegends.com/cdn/11.11.1/img/champion/Aatrox.png. Simply substitute the champion name in this link will give a small square image of each character. This important because on our minimap we want to display each person by their specific champion. At some point, I will use a Javascript library or something to circle off each image since, currently, they come in square format.
